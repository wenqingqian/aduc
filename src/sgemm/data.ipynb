```python
import pandas as pd
import matplotlib.pyplot as plt


# DataFrame
df = pd.read_csv('output.txt')

# 
plt.figure(figsize=(10, 6))
for gemm_type in df['name'].unique():
    if not 'Gmem' in gemm_type: continue
    plt.plot(df[df['name'] == gemm_type]['M'], df[df['name'] == gemm_type]['gflops'], marker='.', label=gemm_type)
    
# GFLOPS
special_gflops_value = 16353.280  # GFLOPS
plt.axhline(y=special_gflops_value, color='r', linestyle='--', label=f'Peak GFLOPS Value: {special_gflops_value}')

plt.xlabel('Matrix Size (M)')
plt.ylabel('Performance (GFLOPS)')
plt.title('Performance Comparison of GEMM Implementations')
plt.legend()
plt.grid(True)
plt.show()
```
